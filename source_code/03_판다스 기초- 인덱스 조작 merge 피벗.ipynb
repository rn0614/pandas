```python
#df 생성
df1=pd.DataFrame({
    '고객번호': [1001,1002,1003,1004,1005,1006,1007],
    '이름' :['둘리','도우너','또치','길동','희동','마이콜','영희']})
df1
```

```python
df2= pd.DataFrame({
    '고객번호': [1001,1002,1005,1006,1008,1001],
    '금액' :[10000, 20000, 30000,40000,50000,60000]})
df2
```

```python
pd.merge(df1, df2) # default how= 'inner' : 양쪽 다 있는 자료만 수집
```

```python
pd.merge(df1, df2, how='outer') # default how= 'inner' : 값없으면 NaN
```

```python
pd.merge(df1, df2, how='left') # default how= 'left' : df1기준
```

```python
pd.merge(df1, df2, how='right') # default how= 'right' : df2기준
```

```python
data={
    '도시':['서울','서울','서울','부산','부산','부산','인천','인천'],
    '연도':['2015','2010','2005','2010','2015','2015','2015','2010'],
    '인구':[995959,995959,995959,995959,995959,995959,995959,995959],
    '지역':['수도권','수도권','수도권','경상권','경상권','경상권','수도권','수도권']
    
}
df3=pd.DataFrame(data)
df3
```

```python
df3.pivot('도시','연도','인구')
```

```python
df3.pivot_table('인구', '도시', '연도')
```

```python
df3.pivot_table('인구', '도시', '연도', margins=True, margins_name='평균')
```

```python
df3.pivot_table('인구', index=['연도', '도시'])
```

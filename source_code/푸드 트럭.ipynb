```python
import pandas as pd
import numpy as np
```

```python
food_truck = pd.read_csv('../data/food_truck.csv', encoding='utf-8')
food_truck.head(10)
```

```python
food_truck.columns
```

```python
food_truck.rename(columns={food_truck.columns[0] :'구별'},inplace=True)
```

```python
food_truck2=food_truck[['구별','위도','경도','푸드트럭운영대수']]
food_truck2.head()
```

```python
food_truck2.drop_duplicates(['구별'])
food_truck2['구별'].unique()
```

```python
food_truck2[food_truck2['구별'].isnull()]
```

```python
food_truck2[food_truck2['푸드트럭운영대수'].isnull()]
```

```python
food_truck2=food_truck2.dropna()
```

```python
food_truck2[food_truck2['푸드트럭운영대수'].isnull()]
```

```python
# 정렬된 결과를 파일로 저장 : encoding='euc-kr'
food_truck2.to_csv('../data/food_truck_result.csv',
                      header=True,
                      index=True,
                      encoding='utf-8')
```

```python
food_truck2['푸드트럭운영대수'].value_counts() # 운영대수별 지역 수
```

```python
import seaborn
import folium
```

```python
food_truck2 = pd.read_csv('../data/food_truck_result.csv', index_col='구별')
food_truck2.head(10) # 인덱스가 9에서 끊겨서 아래의 지도 표시에서 반복문 실행시 오류
```

```python
food_truck4=food_truck2
food_truck4.head(145)
```

```python
food_truck4.index
```

```python
food_truck_map = folium.Map(location=[37.566345, 126.977893], zoom_start=11)

for name in food_truck4.index:
    folium.Marker([food_truck4['위도'][name], food_truck4['경도'][name]], popup=name).add_to(food_truck_map)


food_truck_map.save('../data/food_truck_map.html')
food_truck_map
```
